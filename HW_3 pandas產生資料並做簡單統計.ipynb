用pandas產生dataframe並做簡單統計分析

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

teams = list("ABCDEFG")
score = [60, 70, 55, 80, 90, 95, 65]
data_dict = {'teams': teams,'score': score}

score_df = pd.DataFrame(data_dict)
print(score_df)
```

```python
#敘述統計
score_df.describe()

#存檔為csv
score_df.to_csv('data.csv')
```

```python
#折線圖
score_df.plot()

#散點圖
plt.scatter(teams,score,color='b',label='考試資料')
plt.xlabel('Teams')
plt.ylabel('Score')
plt.show()
```
