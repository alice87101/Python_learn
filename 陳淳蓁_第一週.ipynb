請依照本週的範圍（變數設定、基本運算、print......），自己出一題並完成！
（可以從自己的生活或研究領域出題）

```python
#計算BMI值並且判斷身體狀態

h=float(input('請輸入你的身高(公尺):'))
w=float(input('請輸入你的體重(公斤):'))
B=round(w/h**2, 2)

if B<=15:
    print("你的BMI值為",B,"身體狀態為非常嚴重的體重不足。")
elif 15<B<=16:
    print("你的BMI值為",B,"身體狀態為嚴重體重不足。")
elif 16<B<=18.5:
    print("你的BMI值為",B,"身體狀態為體重過輕。")
elif 18.5<B<=25:
    print("你的BMI值為",B,"身體狀態為體重正常(健康體重)。")
elif 25<B<=30:
    print("你的BMI值為",B,"身體狀態為體重過重。")
elif 30<B<=35:
    print("你的BMI值為",B,"身體狀態為肥胖I級(中等肥胖)。")
elif 35<B<=40:
    print("你的BMI值為",B,"身體狀態為肥胖2級(嚴重肥胖)。")
elif B>40:
    print("你的BMI值為",B,"身體狀態為肥胖3級(非常嚴重肥胖)。")
else:
    print("無法判斷")
```
