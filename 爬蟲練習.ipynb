```python
import requests
from bs4 import BeautifulSoup

tag = input("class.id# ")
res = requests.get('https://www.ncbi.nlm.nih.gov/pubmed/?term=Schizophrenia')
soup = BeautifulSoup(res.text, "lxml")

for drink in soup.select('{}'.format(tag)):
    print(drink.get_text())

```
